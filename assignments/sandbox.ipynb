```python
import pandas as pd
```

```python
df = pd.read_csv('life_expectancy/data/eu_life_expectancy_raw.tsv',sep='\t')
print(df.head())
```

```python
df[['unit', 'sex', 'age', 'region']] = df['unit,sex,age,geo\\time'].str.split(',', expand=True)

df.drop('unit,sex,age,geo\\time', axis=1, inplace=True)

df = pd.melt(df, id_vars=['unit', 'sex', 'age', 'region'], var_name="year")

df['value'].replace(': ', None, inplace=True)
df['value'] = df['value'].str.split(' ', expand=True)[0]

df['year'] = df['year'].astype(int)
df['value'] = df['value'].astype(float)

df = df.dropna()
df = df[df["region"] == "PT"]

df.to_csv('life_expectancy/data/pt_life_expectancy.csv')

print(df)
```
